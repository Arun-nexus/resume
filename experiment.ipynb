```python
import pandas as pd
df3 = pd.read_csv(r"C:\Users\Arun\Downloads\candidate_job_role_dataset.csv")
```

```python
df2.drop(columns=["job_id","category","job_description"],inplace= True)
```

```python
import ast

def removing(df):
    df["job_skill_set"] = df["job_skill_set"].apply(
        lambda x: ", ".join(ast.literal_eval(x)) if isinstance(x, str) else ", ".join(x)
    )
    return df

df2 = removing(df2)
```

```python
df2
```

```python
df3.drop(columns=["candidate_id","qualification","experience_level"],inplace= True)
df3
```

```python
skills = pd.concat((df3["skills"],df2["job_skill_set"]),ignore_index=True)
title = pd.concat((df3["job_role"],df2["job_title"]),ignore_index=True)
```

```python
df = {"skills":skills,
      "title":title}

df = pd.DataFrame(df)
```

```python
df.dropna(axis=1,inplace= True)
df.drop_duplicates(inplace= True)
```

```python
df.to_csv("data/dataset.csv",index = False)
```

```python
from sklearn.preprocessing import OrdinalEncoder , LabelEncoder
import numpy as np
```

```python
title_counts = df['title'].value_counts()
valid_titles = title_counts[title_counts >= 5].index
df_filtered = df[df['title'].isin(valid_titles)].copy()
```

```python
df_filtered
```

```python
df_filtered.to_csv("data/dataset.csv",index=False)
```

```python
ordinal = OrdinalEncoder()
label = LabelEncoder()
ordinal_title = ordinal.fit_transform(np.array(df_filtered["title"]).reshape(-1,1))
lable_title = label.fit_transform(df_filtered["title"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(tokenizer= lambda x :[i.strip().lower() for i in x.split(',')],token_pattern= None)
tfidf_skills = tfidf.fit_transform(df_filtered["skills"])
```

```python
tfidf_skills = tfidf_skills.toarray()
```

```python
tfidf_skills
```

```python
from sklearn.decomposition import TruncatedSVD
components = [50,100,200,300,400]
truncated_data = []
for i in components:
    svd = TruncatedSVD(n_components=i,random_state = 42)
    skills_reduce = svd.fit_transform(tfidf_skills)
    truncated_data.append(skills_reduce)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
```

```python
model_for_normal = XGBClassifier(
    n_jobs=-1,
    eval_metric='mlogloss',
    tree_method='hist',
    device='cuda'
)
model_for_svd0 =  XGBClassifier(

    n_jobs=-1,
    eval_metric='mlogloss',
    tree_method='hist',
    device='cuda'
)

model_for_svd1 =  XGBClassifier(

    n_jobs=-1,
    eval_metric='mlogloss',
    tree_method='hist',
    device='cuda'
)

model_for_svd2 =  XGBClassifier(

    n_jobs=-1,
    eval_metric='mlogloss',
    tree_method='hist',
    device='cuda'
)

model_for_svd3 =  XGBClassifier(

    n_jobs=-1,
    eval_metric='mlogloss',
    tree_method='hist',
    device='cuda'
)

model_for_svd4 =  XGBClassifier(

    n_jobs=-1,
    eval_metric='mlogloss',
    tree_method='hist',
    device='cuda'
)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(tfidf_skills,ordinal_title,test_size=0.3)
x_train_svd0,x_test_svd0,y_train_svd0,y_test_svd0 = train_test_split(truncated_data[0],ordinal_title,test_size=0.3)
x_train_svd1,x_test_svd1,y_train_svd1,y_test_svd1 = train_test_split(truncated_data[1],ordinal_title,test_size=0.3)
x_train_svd2,x_test_svd2,y_train_svd2,y_test_svd2 = train_test_split(truncated_data[2],ordinal_title,test_size=0.3)
x_train_svd3,x_test_svd3,y_train_svd3,y_test_svd3 = train_test_split(truncated_data[3],ordinal_title,test_size=0.3)
x_train_svd4,x_test_svd4,y_train_svd4,y_test_svd4 = train_test_split(truncated_data[4],ordinal_title,test_size=0.3)
```

```python
model_for_normal.fit(x_train,y_train)
```

```python
model_for_svd0.fit(x_train_svd0,y_train_svd0)
model_for_svd1.fit(x_train_svd1,y_train_svd0)
model_for_svd2.fit(x_train_svd2,y_train_svd2)
model_for_svd3.fit(x_train_svd3,y_train_svd2)
model_for_svd4.fit(x_train_svd4,y_train_svd4)
```

```python
from sentence_transformers import SentenceTransformer,util
from sklearn.metrics.pairwise import cosine_similarity
```

```python
transformer = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
df_filtered = df_filtered.reset_index(drop=True)
```

```python
skills_embedding = transformer.encode(df_filtered["skills"])
label_embedding = transformer.encode(df_filtered["title"])
```

```python
sentence_cosine_similarity = util.cos_sim(skills_embedding,label_embedding)
```

```python
sklearn_cosine = cosine_similarity(skills_embedding,label_embedding)
```

```python
import numpy as np
print(np.allclose(sentence_cosine_similarity, sklearn_cosine, atol=1e-5))
print("Max difference:", np.abs(sentence_cosine_similarity - sklearn_cosine).max())
```
